```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn
import pandas as pd
from datetime import date
seaborn.set()
pd.plotting.register_matplotlib_converters()
```

```python
from parse_feedback import read_txt

feedback = read_txt()
```

```python
feedback
```

```python
overall = feedback[feedback['Category'] == 'Overall'].copy()

ranking_type = pd.api.types.CategoricalDtype(['Difficulty in most areas',
                                              'Will probably meet in some areas and not in others',
                                              'Will meet or exceed in most areas', 
                                              'Will probably exceed expectations in most areas',
                                              'Outstanding - will exceed in all areas',
                                              ''], ordered=True)

overall['Ranking'] = overall['Ranking'].astype(ranking_type)
ax = seaborn.stripplot(x='Time', hue='Affiliation', y='Ranking', jitter=True, dodge=True, data=overall)
ax.figure.set_size_inches(8, 4)
ax.set_xlim(date(2019,4,12), date(2019,5,1))
plt.legend(bbox_to_anchor=(1.01, 1), loc=2, borderaxespad=0.)
for item in ax.get_xticklabels():
    item.set_rotation(45)
```
